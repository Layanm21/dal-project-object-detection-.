```python
from pathlib import Path
import random, shutil

root = Path('/content/dal')
img_tr = root/'images/train'
lbl_tr = root/'labels/train'
img_val = root/'images/val'
lbl_val = root/'labels/val'

img_val.mkdir(parents=True, exist_ok=True)
lbl_val.mkdir(parents=True, exist_ok=True)

imgs = sorted([p for p in img_tr.glob('*.jpg')])
random.seed(42)
n_val = int(len(imgs)*0.15)
val_set = set(random.sample([p.stem for p in imgs], n_val))

for stem in val_set:
    for ext in ['.jpg','.jpeg','.png']:
        s = img_tr/f'{stem}{ext}'
        if s.exists():
            shutil.move(str(s), str(img_val/s.name))
            break
    t = lbl_tr/f'{stem}.txt'
    if t.exists():
        shutil.move(str(t), str(lbl_val/t.name))

print('val images:', len(list(img_val.glob('*'))))
```

```python
# 2) أنشئ val حقيقي 15% من train (مو symlink)
from pathlib import Path
import random, shutil

root = Path('/content/dal')
img_tr, lbl_tr = root/'images/train', root/'labels/train'
img_val, lbl_val = root/'images/val', root/'labels/val'
img_val.mkdir(parents=True, exist_ok=True)
lbl_val.mkdir(parents=True, exist_ok=True)

imgs = sorted([p for p in img_tr.glob('*.jpg')])
random.seed(42)
val_n = int(len(imgs)*0.15)
val_stems = set(p.stem for p in random.sample(imgs, val_n))

for stem in val_stems:
    moved = False
    for ext in ['.jpg','.jpeg','.png']:
        s = img_tr/f'{stem}{ext}'
        if s.exists():
            shutil.move(str(s), str(img_val/s.name))
            moved = True
            break
    t = lbl_tr/f'{stem}.txt'
    if t.exists():
        shutil.move(str(t), str(lbl_val/t.name))

print("train images:", len(list(img_tr.glob('*.jpg'))))
print("val images:", len(list(img_val.glob('*.jpg'))))
```

```python
import pandas as pd

inp = '/content/dal/submission_final.csv'
out = '/content/dal/submission_final_clean.csv'

df = pd.read_csv(inp)

def dedup_pred(s):
    if pd.isna(s) or not str(s).strip():
        return ""
    lines = [x.strip() for x in str(s).splitlines() if x.strip()]
    uniq = list(dict.fromkeys(lines))
    return "\n".join(uniq)

df["predictions"] = df["predictions"].apply(dedup_pred)
df.to_csv(out, index=False)
print("saved:", out)
```
